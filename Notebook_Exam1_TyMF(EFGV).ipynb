**(a)** Constant volume process (isometric or isochoric process).

*El proceso de volumen constante, también conocido como proceso isométrico o isocórico, es aquel en el cual el volumen del sistema permanece constante mientras ocurre un cambio en su estado termodinámico. En este proceso, la energía transferida al sistema se convierte en aumento de temperatura y, por lo tanto, en aumento de su energía interna.*

*Durante un proceso isométrico, el volumen V del sistema permanece constante, lo que implica que el trabajo realizado W = 0. Por lo tanto, la primera ley de la termodinámica para este proceso queda expresada como:*
$$\Delta  U = Q$$

*donde $\Delta U$ es el cambio en la energía interna del sistema y Q es la cantidad de calor transferida al sistema.*

*Durante un proceso isométrico, el sistema no realiza trabajo, ya que la presión es constante. Sin embargo, se puede transferir calor al sistema, lo que produce un aumento en su temperatura y en su energía interna.*

*Donde $\Delta S es el cambio en la entropia del sistema y T es la Temperatura absoluta$*

- ### **Ejemplo:**

*Simulacion de un proceso isometrico o isocorico, graficar la  presion  y la temperatura e un gas ideal en funcion del tiempo mientras se mantiene el volumen constante,datos (1 mol de gas ideal,  temperatura inicial de 300 K , presion inicial de 1 atom, tiempo de 10 segundos*




```python
import matplotlib.pyplot as plt
import numpy as np

# Datos iniciales
n = 1 # mol
V = 1 # m^3
R = 8.314 # J/mol*K
T0 = 300 # K
P0 = 1 # atm
t_final = 10 # segundos

# Definir la ecuación de estado del gas ideal
def gas_ideal(P, V, n, R, T):
    return (n*R*T) / V

# Definir el tiempo y la temperatura inicial
t = np.linspace(0, t_final, 1000)
T = np.ones_like(t) * T0

# Calcular la presión en función del tiempo utilizando la ecuación de estado
P = gas_ideal(P0, V, n, R, T)

# Graficar la presión y la temperatura en función del tiempo
fig, ax1 = plt.subplots()

color = 'tab:red'
ax1.set_xlabel('Tiempo (s)')
ax1.set_ylabel('Presión (atm)', color=color)
ax1.plot(t, P, color=color)
ax1.tick_params(axis='y', labelcolor=color)

ax2 = ax1.twinx()

color = 'tab:blue'
ax2.set_ylabel('Temperatura (K)', color=color)
ax2.plot(t, T, color=color)
ax2.tick_params(axis='y', labelcolor=color)

fig.tight_layout()
plt.show()


```

*La grafica muestra la presión en función del tiempo en rojo y la temperatura en función del tiempo en azul. Como el volumen del gas se mantiene constante, la presión del gas aumenta proporcionalmente a la temperatura. En otras palabras, la presión y la temperatura del gas se mantienen en una relación constante conocida como la ley de Charles. En la gráfica, esto se ve como una línea recta con una pendiente positiva para la presión y la temperatura.*

*Es importante tener en cuenta que este es solo un ejemplo simplificado de un proceso isocórico y que en la práctica, puede haber otros factores que afecten la relación entre la presión y la temperatura.*

- ### **(b). Constant pressure process (isopiestic or isobaric process).**

*El proceso de presión constante, también conocido como proceso isobárico o isopéstico, es un proceso termodinámico en el cual la presión del sistema permanece constante mientras que el volumen cambia. La presión dentro del recipiente se mantiene constante mientras el calor se transfiere al alimento, lo que permite que el volumen del aire dentro del recipiente se expanda. Como la presión se mantiene constante, el trabajo realizado por el sistema puede calcularse fácilmente a partir del cambio en volumen.*

*Supongamos que tenemos un gas encerrado en un cilindro con un pistón que mantiene la presión constante. Si el gas se expande al recibir calor, el trabajo realizado por el gas puede calcularse como el producto de la presión constante y la variación del volumen del gas. Además, podemos calcular la variación de la energía interna utilizando la ecuación $\Delta U = Q - W$, donde $Q$ es la cantidad de calor transferida y $W$ es el trabajo realizado por el gas. y los datos dados una presión constante de 1 atm y un volumen inicial de 1 litro a una temperatura de 300 K. El gas se expande a un volumen final de 2 litros al recibir 1000 J de calor. Queremos calcular el trabajo realizado por el gas y graficar la variación de la energía interna con la temperatura.*



```python
import numpy as np
import matplotlib.pyplot as plt

# Datos del problema
P = 1.0     # Presión constante (en atm)
V1 = 1.0    # Volumen inicial (en litros)
V2 = 2.0    # Volumen final (en litros)
Q = 1000.0  # Cantidad de calor transferida (en Joules)
T = np.linspace(300, 500, 100)  # Temperaturas a graficar (en Kelvin)

# Cálculo del trabajo
W = P * (V2 - V1)  # Trabajo realizado por el gas (en Joules)

# Cálculo de la variación de la energía interna
delta_U = Q - W  # Variación de la energía interna (en Joules)

# Gráfica de la variación de la energía interna con la temperatura
U = delta_U + (3/2)*n*R*T
plt.plot(T, U, color='red')
plt.xlabel('Temperatura (K)')
plt.ylabel('Energía interna (J)')
plt.title('Variación de la energía interna')
plt.show()
```

- ### **(c) Constant temperature process (or isothermal process or hyperbolic process).**

*El proceso isotérmico es un proceso termodinámico en el cual la temperatura del sistema permanece constante durante todo el proceso. En otras palabras, la temperatura no varía aunque se le transfiera calor al sistema o se realice trabajo sobre él. Esto implica que la energía interna del sistema no cambia. el proceso isotermico se puede representar constante.*

$$P \cdot V = n \cdot R \cdot T$$

*donde P es la presión, V es el volumen, n es la cantidad de sustancia, R es la constante universal de los gases y T es la temperatura constante del proceso.*

*Físicamente, en un proceso isotérmico se puede demostrar que la presión y el volumen de un gas son inversamente proporcionales, lo que se conoce como la ley de Boyle-Mariotte. Esto significa que cuando la temperatura se mantiene constante, si la presión aumenta, el volumen del gas disminuye y viceversa. De esto tambien podemos dar con el trabajo realizado por un gas se puede  carcular mediante la formula.*

$$ W = n \cdot R \cdot T \cdot ln \left ( \frac{V_f}{V_i} \right ) $$

*Donde W es el trabajo realizado, n es la cantidad de sustancia, R es la constante universal de los gases, T es la temperatura constante del proceso, Vi es el volumen inicial y Vf es el volumen final.*

*El cambio en la energía interna en un proceso isotérmico es cero, ya que la temperatura se mantiene constante y la energía interna de un gas depende únicamente de su temperatura.*



- ### ***Trabajo relizado por un Gas:***

*En un proceso isotérmico, la temperatura del sistema se mantiene constante, lo que significa que la energía interna del gas no cambia. El trabajo realizado por un gas en un proceso isotérmico puede ser calculado mediante la fórmula:*

- ### ***Cambio en la Energia  Interna:***

*En un proceso isotérmico, la temperatura del sistema se mantiene constante, lo que significa que la energía interna del gas no cambia. Por lo tanto, el cambio en la energía interna (ΔU) del gas durante un proceso isotérmico es cero. Esto se debe a que la energía térmica transferida al gas en forma de calor se utiliza completamente para realizar trabajo, y no hay cambio en la energía interna del gas.*

*Podemos demostrar esto utilizando la Primera Ley de la Termodinámica, que establece que la variación de la energía interna de un sistema es igual a la cantidad de calor que entra en el sistema menos el trabajo que sale del sistema:*

$$\Delta U = Q-W$$

*Por lo tanto, el cambio en la energía interna durante un proceso isotérmico es cero, ya que toda la energía térmica transferida al gas en forma de calor se utiliza para realizar trabajo.*

*Donde $m$ representa  la mas, y $c$ el calor especifico.*

- ### ***Tranferencia de Calor:***
*En un proceso isotérmico, la temperatura del sistema se mantiene constante, lo que significa que no hay un gradiente de temperatura para que se produzca una transferencia de calor. Por lo tanto, en un proceso isotérmico ideal, no hay transferencia de calor entre el gas y su entorno.*

*Sin embargo, en la práctica, es difícil mantener una temperatura constante perfectamente, por lo que es posible que se produzcan pequeñas transferencias de calor durante un proceso isotérmico real. En este caso, la transferencia de calor se puede calcular utilizando la Ley de Fourier de conducción de calor:*
$$\frac {Q}{t} = -kA \left (\frac {dT}{dx} \right)$$

*Donde $Q/t$ es la cantidad de calor transferida por unidad de tiempo, k es la conductividad térmica del material, A es el área de transferencia de calor, $dT/d$ es el gradiente de temperatura y x es la distancia a lo largo de la dirección del gradiente de temperatura.*

*En un proceso isotérmico ideal, el gradiente de temperatura es cero, por lo que la cantidad de calor transferida (Q/t) es cero. En resumen, en un proceso isotérmico ideal, no hay transferencia de calor entre el gas y su entorno, pero en un proceso isotérmico real, puede haber pequeñas transferencias de calor debido a las imperfecciones en el aislamiento térmico del sistema.*


*A lo deduccion aplicando la primera  ley  de la termodinamica tenemnos que la escuacion quda reescrita , como:*

$$\Delta  Q = dU + \Delta W$$

$$\int_{1}^{2} \Delta Q =\int_{1}^{2} \Delta W$$

$$Q_{1-2}= W_{1-2} $$


*Conociendo que la energia interna en este cao es o, esto queda como:*

- ### ***Cambio en la Entalpía:***

*En un proceso isotérmico ideal, la temperatura del sistema se mantiene constante, lo que significa que la energía térmica total del sistema permanece constante. $$H_2 - H_1 = mc_v (T_2 - T_1) + (mRT_2 - mRT_1)$$

$$H_2 - H_1 = mc_v (T_2 - T_1) + mR (T_2 - T_1)$$

$$ H_2 - H_1= m(T_2 - T_1)(c_v + R)$$
*como sabemos  que $R=c_p - c$ entnces*

$H_2 - H_1 = mc_p (T-_2 - T_1)$

*Por lo tanto la entalpia es 0, ya que no vemos ningun cambio en la temperatura*

```python
import matplotlib.pyplot as plt
import numpy as np

# Datos iniciales
n = 1.0   # Número de moles del gas
T = 300  # Temperatura en Kelvin
R = 8.31 # Constante de los gases ideales

# Volumen inicial y final
V1 = 1
V2 = 4

# Cálculo de la presión inicial
P1 = n * R * T / V1

# Función para calcular la presión en función del volumen
def pressure(V):
    return n * R * T / V

# Creación de un array de volumenes
V = np.linspace(V1, V2, 100)

# Cálculo de la presión para cada volumen
P = pressure(V)

# Gráfica de la variación de la presión con el volumen
plt.plot(V, P)
plt.xlabel('Volumen (m^3)')
plt.ylabel('Presión (Pa)')
plt.title('Proceso isotérmico a {} K'.format(T))
plt.show()
```

### **(d) Adiabatic process (or Isentropic process)**

*Un proceso adiabático, también conocido como proceso isentrópico, es aquel en el cual no hay transferencia de calor hacia o desde el sistema en cuestión. En otras palabras, la energía térmica del sistema se mantiene constante durante el proceso. Esto se logra mediante un aislamiento térmico adecuado del sistema, de modo que no haya transferencia de calor a través de sus fronteras.*

*Matemáticamente, podemos describir un proceso adiabático utilizando la primera ley de la termodinámica:*

$$\Delta U = Q-W$$

*donde ΔU es el cambio en la energía interna del sistema, Q es la cantidad de calor transferido hacia o desde el sistema y W es el trabajo realizado por el sistema. En un proceso adiabático, Q es igual a cero, lo que significa que todo el cambio en la energía interna del sistema se debe al trabajo realizado por el sistema.*

*En un proceso adiabático, la entropía del sistema se mantiene constante. La entropía es una medida de la cantidad de desorden o aleatoriedad presente en un sistema. La segunda ley de la termodinámica establece que la entropía de un sistema cerrado nunca disminuye, lo que significa que un proceso adiabático es también un proceso isentrópico, es decir, un proceso en el que no hay cambio en la entropía del sistema.*

*Para un gas ideal, podemos utilizar la ley de los gases ideales para describir el proceso adiabático. La ley de los gases ideales establece que la presión, el volumen y la temperatura de un gas están relacionados por la ecuación:*
$$PV = nRT$$

*Donde P es la presión del gas, V es su volumen, n es el número de moles de gas presente, R es la constante de los gases ideales y T es su temperatura absoluta.*


- ### **Trabajo realizado por un gas:**

*En un proceso adiabático, el trabajo realizado por un gas se debe a una variación en su volumen. Debido a que el proceso es adiabático, no hay transferencia de calor hacia o desde el sistema, lo que significa que todo el cambio en la energía interna del gas se debe al trabajo realizado por el gas.*

*En un proceso adiabático, el trabajo realizado por el gas se puede calcular utilizando la relación adiabática para un gas ideal. Esta relación establece que la presión y el volumen del gas están inversamente relacionados cuando la cantidad de gas y la temperatura se mantienen constantes. En otras palabras, si el volumen del gas disminuye, su presión aumenta, y viceversa.*

*Imaginemos un pistón con un gas dentro que se encuentra en un cilindro con un émbolo que se puede mover hacia adentro y hacia afuera. Si el émbolo se mueve hacia adentro, el volumen del gas disminuye y su presión aumenta. Como resultado, el gas realiza trabajo sobre el émbolo y la energía del gas disminuye.*

*Para un proceso adiabático en un gas ideal, podemos utilizar la relación adiabática, que establece que:*

$$PV^y =  constante$$

*Donde γ es la relación de calor específico del gas (es decir, la razón entre el calor específico a presión constante y el calor específico a volumen constante). Esta relación adiabática nos permite calcular la variación de presión y volumen en el proceso adiabático.*




*Este caso para el trabajo de un debemos tener en cuenta que  la fotrmul a trabajar es respecto a $P$.  $V^y$  como $k^y$, y $W  \int_{1}^{2} p dV$*

$$\Delta W =\int_{1}^{2}  \frac{k}{V^y} dv = \frac{K}{1-Y}\left ( \frac{1}{V_{2}^{y-1}}-\frac{1}{V_{2}^{y-1}} \right ) =\frac{1}{1-y}\left ( \frac{P_2 \cdot P_{2}^{y}}{V_{2}^{y-1}}- \frac{P_1 \cdot V_{1}^{y}}{V_{1}^{y}-1} \right )= \frac{1}{1-y} (P_2 \cdot V_2 - P_1 \cdot V_1) =\frac{P_2 \cdot V_2 - P_1 \cdot V_1}{1-y}$$


- ### **Cambio en la energia interna:**
*En un proceso adiabático, el cambio interno en el gas se debe a la realización de trabajo por parte del gas sin transferencia de calor hacia o desde el sistema. Esto significa que el cambio interno en el gas está relacionado directamente con el trabajo realizado por el gas.*

*Cuando el gas realiza trabajo en un proceso adiabático, su energía interna disminuye. Esto se debe a que el trabajo realizado por el gas se utiliza para vencer la resistencia del entorno y reducir el volumen del gas. Al disminuir el volumen del gas, la energía interna del gas también disminuye, ya que la energía cinética de las moléculas del gas se reduce debido a la disminución de la distancia recorrida por las moléculas.*

*Es importante tener en cuenta que en un proceso adiabático, la energía interna del gas solo cambia debido al trabajo realizado por el gas. Si el proceso fuera no adiabático, es decir, si hubiera transferencia de calor hacia o desde el sistema, entonces el cambio en la energía interna del gas también estaría influenciado por la transferencia de calor.*

*La variación de la energía interna en un proceso adiabático puede expresarse matemáticamente como:*

$$\Delta U = W$$

*Donde $\Delta U$ es la variación de la energía interna del gas y W es el trabajo realizado por el gas. Para un gas ideal, la relación adiabática establece que la presión y el volumen del gas están inversamente relacionados cuando la cantidad de gas y la temperatura se mantienen constantes. Por lo tanto, el trabajo realizado por el gas en un proceso adiabático.*

*llevandola a trabajar con la primera ley de la termodinamica ,el cambio en la energia interna se rige por $dU = mc_v dT$, conociendo la "m"   que representa la masa, "c" el calor especifico, con esto defini,  integramos la ecuacion.*

$$\int_{1}^{2}dU =\int_{1}^{2} mc_v dT = mc_v \int_{1}^{2}dT$$

$$ U_2 - U_1 = mc_v (T_2 - T_1) $$


- ### **Transferencia de Calor:**

*En un proceso adiabático, no hay transferencia de calor hacia o desde el sistema, ya que el sistema está aislado térmicamente del entorno. Esto significa que la cantidad de calor que entra o sale del sistema es cero, y la energía se transfiere solo en forma de trabajo.*

*La ausencia de transferencia de calor en un proceso adiabático se debe a la presencia de un aislante térmico que impide el flujo de calor entre el sistema y el entorno. Como resultado, cualquier cambio en la energía interna del sistema solo puede ocurrir debido a la realización de trabajo en el sistema o debido a la variación de la energía cinética o potencial de las moléculas en el sistema.*

*Es importante tener en cuenta que la ausencia de transferencia de calor en un proceso adiabático no significa que no pueda haber cambios en la temperatura del sistema. La temperatura del sistema puede cambiar debido a la realización de trabajo, que puede causar una disminución de la energía interna del sistema y, por lo tanto, una disminución de la temperatura. Sin embargo, estos cambios de temperatura se deben exclusivamente al trabajo realizado y no a la transferencia de calor.*

$$ \Delta Q = dU + \Delta W$$

*Integrando la ecuacion  de transferencia de calor dada por la primera ley de la termodinamica, tenieendo en cuenta que lo hacemos en desplazamiento entre los  dos estados.*

$$\int_{1}^{2}\Delta Q = \int_{1}^{2} dU +\int_{1}^{2} \Delta W$$

$$Q_{1-2}=(U_2 - U_1) + W_{1-2}$$

*pero conocemos que ene ste caso  $W= -\Delta U$  entonces  nuestra transferencia de calor será igual a 0*


- ### **Cambio en la Entalpia:**

*En un proceso adiabático, no hay transferencia de calor hacia o desde el sistema, por lo que la variación de la entalpía del sistema está relacionada solo con el trabajo realizado sobre el sistema.*

*La entalpía (H) es una función termodinámica que describe la cantidad de energía que se encuentra en un sistema a presión constante. En un proceso adiabático, la presión puede cambiar, pero se mantiene constante la presión externa, que es la presión a la que el sistema está expuesto.*

*La variación de la entalpía $(\Delta H)$ en un proceso adiabático se puede calcular utilizando la siguiente fórmula:*

$$  \Delta H = \Delta U + P \Delta V$$

*donde $ΔU$ es la variación de la energía interna del sistema y ΔV es la variación de su volumen. En un proceso adiabático, la cantidad de calor transferida (Q) es cero, por lo que $ΔU = W$, donde W es el trabajo realizado sobre el sistema. Por lo tanto, la fórmula se puede reescribir como:*

$$\Delta H = W + P \Delta V$$

*En un proceso adiabático, el trabajo realizado sobre el sistema es igual al cambio en la energía interna del sistema, por lo que la fórmula se puede simplificar aún más como:*

$$\Delta H = \Delta U  + P \Delta V = 0 +  P \Delta V = P \Delta V$$

*Por lo tanto, en un proceso adiabático, la variación de la entalpía del sistema es igual al producto de la presión del sistema y la variación de su volumen.*

*Reescribiendo la  ecuacion  en forma de cambio en su entalpia.*
$$ dH = dU + d(pV)$$


*Conociendo que  $U_1 - U_2 = mc_v(T_2 - T_1) : P_1 V_1  mRT_1 : P_2 V_2= mRT_2$ entonces*

$$H_2 - H_1 = mc_v (T_2 - T_1) + (mRT_2 - mRT_1)$$

$$H_2 - H_1 = mc_v (T_2 - T_1) + mR (T_2 - T_1)$$

$$ H_2 - H_1= m(T_2 - T_1)(c_v + R)$$
*como sabemos  que $R=c_p - c$ entnces*

$H_2 - H_1 = mc_p (T-_2 - T_1)$



- ### **Ejemplo:**

Simular un proceso adiabatico






```python
import numpy as np
import matplotlib.pyplot as plt

# Parámetros iniciales del sistema
T0 = 300  # temperatura inicial en K
P0 = 1  # presión inicial en atm
V0 = 1  # volumen inicial en m^3
gamma = 1.4  # relación de calor específico del gas

# Cálculo de la compresión adiabática
Vf = 0.5 * V0  # volumen final en m^3
Pf = P0 * (V0/Vf)**gamma  # presión final en atm
n = P0*V0/(8.31*T0)  # cantidad de gas en moles
Cv = 3/2*n*8.31/(n*8.31-8.31)  # capacidad calorífica a volumen constante

dU = (Pf*Vf - P0*V0)/(gamma-1)  # variación de la energía interna
dT = -dU / (Cv * n)  # variación de la temperatura
dH = dU + P0*(V0-Vf)  # variación de la entalpía

# Crear un arreglo con los valores de volumen durante el proceso
V = np.linspace(V0, Vf, 100)

# Calcular la variación de presión durante el proceso
P = P0 * (V0/V)**gamma

# Crear un arreglo con los valores de temperatura durante el proceso
T = T0 + dT

# Graficar los resultados
plt.plot(V, P)
plt.xlabel('Volumen (m^3)')
plt.ylabel('Presión (atm)')
plt.show()

```

*la gráfica muestra cómo varía la presión del gas en función del volumen durante un proceso adiabático. En el eje horizontal se encuentra el volumen del gas, expresado en metros cúbicos, y en el eje vertical se encuentra la presión del gas, expresada en atmósferas.*

*En este ejemplo, el gas comienza en un estado inicial con un volumen de 1 metro cúbico y una presión de 1 atmósfera. Durante el proceso adiabático, el gas se comprime adiabáticamente hasta alcanzar un volumen final de 0.5 metros cúbicos.*

*La línea curva muestra cómo la presión del gas cambia durante este proceso. Como se puede ver, la presión del gas aumenta significativamente a medida que el volumen disminuye, lo cual es característico de una compresión adiabática. En este caso, la presión final del gas es de aproximadamente 2 atmósferas.*

- ### **(E). Polytropic process.**

*Un proceso politrópico es aquel en el que la relación entre la presión y el volumen de un gas sigue la ley:*

$$P\cdot V^2 = constante$$
*Donde P es la presión, V es el volumen y n es un exponente que depende de las condiciones del proceso.*

*En términos físicos, un proceso politrópico es aquel en el que la energía interna de un gas se conserva. Es decir, el calor que se agrega o se retira del gas se utiliza completamente para cambiar la presión y el volumen del gas, sin cambiar su energía interna. Este tipo de procesos son importantes en la termodinámica, ya que muchos procesos reales se pueden aproximar como procesos politrópicos.*

*La constante de la ley politrópica puede escribirse como:*

$$ K = P_1 \cdot V_1^n =  P_2 \cdot V_2^n $$



- ### **Trabajo realizado por un gas:**

*El trabajo realizado por un gas en un proceso politrópico es una medida de la energía transferida por el gas durante el proceso. Un proceso politrópico es aquel en el que la presión y el volumen del gas:*

$$PV^2  = C$$

*Donde $P$ es la presión, $V$ es el volumen y $n$ es un exponente que depende de las propiedades del gas y de las condiciones del proceso.*

*El trabajo realizado por el gas durante un proceso politrópico se puede calcular mediante la integral:*

$$ W= \int_{V_1}^{V_2} P dV$$

*Donde $W$ es el trabajo, $P$ es la presión y $dV$ es el cambio infinitesimal en el volumen del gas. Sustituyendo la ecuación politrópica, la integral se puede reescribir como:*

$$W = \frac  {P_2 V_2 - P_1 V_1}{n-1}$$

*Donde $P_1$ y $V_1$ son la presión y el volumen iniciales, y $P_2$ y $V_2$ son la presión y el volumen finales del gas.*


- ### **Cambio en la Energia interna:**

*El cambio en la energía interna de un gas en un proceso politrópico depende de las condiciones específicas del proceso y de las propiedades termodinámicas del gas. La energía interna es una medida de la energía total contenida en un sistema, incluyendo la energía cinética y la energía potencial de sus moléculas. En un proceso politrópico, el cambio en la energía interna del gas se puede calcular utilizando la primera ley de la termodinámica, que establece que la energía total en un sistema cerrado se conserva:*

$$\Delta U = Q-W$$

*Donde $\Delta U$ es el cambio en la energía interna, $Q$ es la cantidad de calor transferido al sistema y $W$ es el trabajo realizado por el gas durante el proceso.*

$$\Delta U = -W$$

*Por lo tanto, el cambio en la energía interna en un proceso politrópico adiabático depende únicamente del trabajo realizado por el gas. Si el proceso no es adiabático, entonces el cambio en la energía interna también dependerá de la cantidad de calor transferido al gas.*

*Redefiniendo los terminos en forma  de que podamos integrarla entre los 2 espacios que venimos trabajando.*

$$\int_{1}^{2} dU = \int_{1}^{2} mc_v dT = mc_V \int_{1}^{2}dT$$

$$U_2 - U_1 = mc_v (T_2 - T_1)$$


- ### **Transferencia de calor:**
*La transferencia de calor en un proceso politrópico depende de las condiciones específicas del proceso y de las propiedades termodinámicas del gas. La transferencia de calor es la transferencia de energía térmica de un cuerpo con mayor temperatura a uno con menor temperatura. En un proceso politrópico, la transferencia de calor puede ocurrir de dos maneras: de forma adiabática o no adiabática.*


*En un proceso politrópico adiabático, la transferencia de calor es cero, lo que significa que no hay transferencia de energía térmica entre el gas y su entorno. Esto se puede lograr a través de un aislamiento térmico del sistema o mediante una transferencia de calor muy rápida que no permita la transferencia de calor significativa durante el proceso.*

*Por otro lado, en un proceso politrópico no adiabático, la transferencia de calor puede ocurrir entre el gas y su entorno. Esto puede suceder si el sistema no está aislado térmicamente o si se realiza trabajo en el gas a través de una fuente externa de calor, como una llama.*

$$\Delta Q = dU +  \Delta W$$

*integrando la ecuacion anterior entre los estados, esta queda* 


$$\int_{1}^{2} \Delta Q = \int_{1}^{2}  dU + \int_{1}^{2} \Delta W$$

$$  \Delta Q = \frac{P_2 \cdot V_2  - P_1 \cdot V_1}{1-n}+ mc_v (T_2 - T_1) $$


- ### **Cambio en la Entalpía:**

*El cambio en la entalpía de un gas en un proceso politrópico depende de las condiciones específicas del proceso y de las propiedades termodinámicas del gas. La entalpía es una función termodinámica que representa la energía total de un sistema a presión constante y se define como la suma de la energía interna del sistema y el producto de la presión y el volumen del sistema:*

$$H = U +PV$$

*En un proceso politrópico, el cambio en la entalpía del gas se puede calcular utilizando la siguiente expresión:*

$$\Delta H = \Delta U + P \Delta V$$

*Donde $\Delta U$ es el cambio en la energía interna del gas y $\Delta V$ es el cambio en el volumen del gas durante el proceso. Si el proceso politrópico es adiabático (sin transferencia de calor), entonces $\Delta U = -W$ (donde $W$ es el trabajo realizado por el gas) y la expresión se puede escribir como:*

$$ dH = dU + d (PV)$$

*Ahora integrando desde un estado al otro*

$$ \int_{1}^{2} dH =\int_{1}^{2}  dU + \int_{1}^{2} d (PV)$$

$$H_2 - H_1 = (U_2 - U_1) + (P_2 \cdot V_2 - P_1 V_1) $$

*Conociendo que  $U_1 - U_2 = mc_v(T_2 - T_1) : P_1 V_1  mRT_1 : P_2 V_2= mRT_2$ entonces*

$$H_2 - H_1 = mc_v (T_2 - T_1) + (mRT_2 - mRT_1)$$

$$H_2 - H_1 = mc_v (T_2 - T_1) + mR (T_2 - T_1)$$

$$ H_2 - H_1= m(T_2 - T_1)(c_v + R)$$
*como sabemos  que $R=c_p - c$ entnces*

$H_2 - H_1 = mc_p (T-_2 - T_1)$



- ### **Ejemplo:**

*Se plante aun prceso politrobico adibatico  de expansion de un gas ideal. tiene valores iniciales como un volumen de 1 metro cúbico y una presión de 100 000 pascales (Pascals). Se desea conocer cómo cambia la presión del gas a medida que este se expande politrópicamente, es decir, a medida que su volumen aumenta de forma proporcional a una potencia del cambio de presión. El índice politrópico adiabático se fija en 1,5.*


```python
from scipy.constants import R
import numpy as np
import matplotlib.pyplot as plt

# Definir las condiciones iniciales del proceso politrópico
V1 = 1.0     # Volumen inicial del gas [m^3]
P1 = 1.0e5  # Presión inicial del gas [Pa]
n = 1.5     # Índice politrópico adiabático
m = 100     # Número de puntos en la simulación

# Calcular la temperatura inicial del gas
T1 = P1 * V1 / (n * R)

# Calcular el trabajo realizado por el gas durante el proceso
W = (P1 * V1 / (n - 1)) * (1 - (1 / (m ** (n - 1))))

# Calcular la presión y el volumen del gas en cada punto del proceso
V = np.linspace(V1, V1 * m, m)
P = P1 * (V1 / V) ** n

# Graficar la presión vs. el volumen del gas durante el proceso politrópico
plt.plot(V, P, '-r')
plt.xlabel('Volumen [m^3]')
plt.ylabel('Presión [Pa]')
plt.title('Proceso politrópico adiabático de expansión')
plt.show()

print("El trabajo realizado por el gas es de", round(W/1000,2), "kJ.")
```

*La gráfica muestra la relación entre el volumen y la presión del gas durante el proceso politrópico adiabático de expansión. En el eje X se representa el volumen del gas en metros cúbicos (m^3) y en el eje Y se representa la presión del gas en pascales (Pa). La curva roja en la gráfica muestra la evolución de la presión del gas a medida que se expande politrópicamente.Se puede ver que la presión del gas disminuye a medida que aumenta el volumen, lo cual es consistente con la ecuación anterior. El proceso polítropico adiabático es representado por una curva que en este caso es de forma exponencial decreciente.*

- ### ***Entalpía:***

*La entalpía es una propiedad termodinámica que se utiliza para describir el estado de un sistema termodinámico en términos de su capacidad para intercambiar calor con el entorno. La entalpía se denota por la letra H y se define como la suma de la energía interna del sistema y el producto de la presión y el volumen del sistema.*

*La entalpía es una función de estado, lo que significa que su valor depende solo del estado inicial y final del sistema y no de cómo se alcanzó ese estado. En términos más simples, la entalpía representa la cantidad de energía térmica almacenada en un sistema termodinámico, incluyendo tanto la energía cinética de las partículas que lo componen como la energía potencial debida a las interacciones entre ellas.*

*La entalpía se utiliza comúnmente en la química para describir el calor liberado o absorbido durante una reacción química. En una reacción exotérmica, la entalpía disminuye, lo que indica que el sistema libera calor al entorno. En una reacción endotérmica, la entalpía aumenta, lo que indica que el sistema absorbe calor del entorno.*

*La entalpía también se utiliza en la ingeniería para describir los procesos termodinámicos que ocurren en las máquinas térmicas, como los motores de combustión interna y las centrales eléctricas.*

$$ U=42 + 3.6pV$$

where $U$  is in $kJ$, $p$  in  $kPa$, and $V$ in cubic metre. If the fluid changes from an initial state of $190 kPa,0.035 m^3$ to a final state of $420 kPa, 0.07 m^3,$ with no work othe thah that donde on the piston, find the direction and magnitude if the work and heat transfer

```python
import sympy as sp

# Definimos las variables y constantes
p, V = sp.symbols('p V')
p1, V1 = 190, 0.035
p2, V2 = 420, 0.07
a = (p2 - p1) / (V2 - V1)
b = p1 - a * V1
U = 42 + 3.6 * p * V

# Definimos la función de presión
p_func = a * V + b

# Calculamos el trabajo
W = -sp.integrate(p_func, (V, V1, V2))

# Calculamos la transferencia de calor
U_inicial = U.subs([(p, p1), (V, V1)])
U_final = U.subs([(p, p2), (V, V2)])
Q = U_final - U_inicial - W

# Imprimimos los resultados
print(f"Trabajo: {W.evalf():.7f} kJ")
print(f"Transferencia de calor: {Q.evalf():.7f} kJ")


```
